```python
# Using One v all classification with Gradient Descent Algorithm for Logistic Regression
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
# extracting the data
train_data = pd.read_csv("train.csv")
train_data.insert(9,'NEW',0)
# making a new feature for giving importance to NaN values
train_data['NEW']=np.where(train_data['Number_Weeks_Used'].isnull(),1,0)
train_data
```

```python
# Replacing NaN values with random values taken from the same column
random_sam=train_data['Number_Weeks_Used'].dropna().sample(train_data['Number_Weeks_Used'].isnull().sum(),random_state=0)
random_sam.index=train_data[train_data['Number_Weeks_Used'].isnull()].index
train_data.loc[train_data['Number_Weeks_Used'].isnull(),'Number_Weeks_Used']=random_sam
# feature scaling to validate the eucleadin distance
train_data['Estimated_Insects_Count']=(train_data['Estimated_Insects_Count']-train_data['Estimated_Insects_Count'].mean())/train_data['Estimated_Insects_Count'].std()
train_data
```

```python
# extracting and making features for training and finding apt weights 
X = np.array(train_data.iloc[:,1:10])
one = np.ones((88858,1),dtype=float)
X = np.matrix(np.hstack((one,X)))
Y1 = np.array([train_data.iloc[:,10]])
Y1 = Y1.T
Y2 = Y1.copy()
Y3 = Y1.copy()
for i in range(88858):
    if Y1[i] == 0:
        Y1[i] = 1
    else:
        Y1[i] = 0
    if Y2[i] == 1:
        pass
    else:
        Y2[i] = 0
    if Y3[i] == 2:
        Y3[i] = 1
    else:
        Y3[i] = 0
Y1=np.matrix(Y1)
Y2=np.matrix(Y2)
Y3=np.matrix(Y3)
#Randomly initilaizing weights
Theta1 = np.matrix(np.ones((10,1)))
Theta2 = np.matrix(np.ones((10,1)))
Theta3 = np.matrix(np.ones((10,1)))
```

```python
def GradDesc(x,y,theta):
    z = x*theta
    h = 1/(1 + np.exp(-z))
    alpha=0.0025
    lamb=100
# minimizing the cost function
    J=(alpha/88858)*(np.transpose(x)*(h-y))
    count=0
    while count<=7000:
    #changing weights to reach the minimum of cost function
        theta=theta*(1-alpha*lamb/88858)-J
        z = x*theta
        h = 1/(1 + np.exp(-z))
        J=(alpha/88858)*(np.transpose(x)*(h-y))
        count+=1
    return theta
```

```python
#Predicting the values
Theta1=GradDesc(X,Y1,Theta1)
z1=X*Theta1
h1=1/(1+np.exp(-z1))
Theta2=GradDesc(X,Y2,Theta2)
z2=X*Theta2
h2=1/(1+np.exp(-z2))
Theta3=GradDesc(X,Y3,Theta3)
z3=X*Theta3
h3=1/(1+np.exp(-z3))
```

```python
# finding the appropriate prediction using one vs all classification method
def maxval(a,b,c):
    if a>=b and a>=c:
        return 0
    if b>=a and b>=c:
        return 1
    if c>=a and c>=b:
        return 2
h = np.matrix(np.zeros((88858,1)))
```

```python
for i in range(88858):
    h[i] = maxval(h1[i][0],h2[i][0],h3[i][0])
y_a=list(train_data['Crop_Damage'])
y_p=list(np.array(h,dtype=int).flatten())
# Confusion Matrix
print("Confusion Matrix : \n",confusion_matrix(y_a,y_p))    
        
```
